```python
import sys

import numpy as np
from matplotlib import pyplot as plt

from src.utils import save_plt

PWD = '../..'
sys.path.append(PWD)

IMAGES_DIR = PWD + '/data/images/useful_plots/'
```

# Sigmoid

```python
plt.figure(figsize=(5, 3))

x = np.linspace(-6, 6, 1000)
y = 1 / (1 + np.exp(-x) )

plt.xlim(-6, 6)
plt.plot(x, y)
plt.tight_layout()

save_plt(IMAGES_DIR + '/sigmoid.pdf')
plt.show()
```

# ReLU

```python
plt.figure(figsize=(5, 3))

x = np.linspace(-6, 6, 1000)
y = np.maximum(0, x)

plt.xlim(-6, 6)
plt.plot(x, y)
plt.tight_layout()

save_plt(IMAGES_DIR + '/relu.pdf')
plt.show()
```
